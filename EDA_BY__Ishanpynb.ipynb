```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("/content/Student_performance_data _.csv")
df.head(5)
```

```python

# Categorical data
# countplot
sns.countplot(data=df, x='Age')
df['Gender'].value_counts()
```

```python
 # pie chart

df['Gender'].value_counts().plot(kind='pie',autopct='%0.2f')
```

```python

# Numberical Value

# Histogram treat as categorical data distribution of data

plt.hist(df['Age'])
```

```python


# Distplot
# probability density function --> The line is kde
# skweness
sns.distplot(df['Age'])
```

```python


# Boxplot
# 5 number summary
# median , percentile, min=Q1-1.5*IQR, max=Q3+1.5*IQR, IQR = (Q1-Q3), outlier

sns.boxplot(df['Age'])
```

**Bivariate** **Analaysis**

```python
# Scatter plot(Numerical - Numerical) for the relation between two column
sns.scatterplot(x='StudentID', y='Age',hue='Gender', style='Ethnicity',size='ParentalEducation' ,data=df.head())
plt.show()
```

```python
  # Bar plot (Numerical - Categorical)
  # average age in pclass
sns.barplot(x='StudentID',y='Age',hue='Gender',data=df.head())
```

```python

# Displot

sns.distplot(df[df['Gender']==0]['Age'],hist=False)
sns.distplot(df[df['Gender']==1]['Age'],hist=False)
```

```python

# HeatMap (Categorical - Categorical)
sns.heatmap(pd.crosstab(df.head()['StudentID'],df.head()['ParentalSupport']))
```
